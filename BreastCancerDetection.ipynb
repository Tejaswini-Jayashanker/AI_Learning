the Aim of this project is to classify the tumors as malignant (cancerous) or benign(non cancerous) using the Breast Cancer Dataset on Kaggle. The unnamed column is dropped and the dataset is cleaned.

This is a Binary classification ML problem.

```python
#import the neccesarry libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import the dataset as Pandas DF
data = pd.read_csv("/content/drive/MyDrive/Machine Learning Projects/DataSets/breast-cancer.csv")
```

```python
print("ANY NULL DATA CHECK:", data.isnull().values.any() )
print( "#NULL DATA CHECK: ", data.isnull().values.sum() )
```

```python
#Encode the Categorical Variables:
data.select_dtypes(include='object').columns #This will be my O/P variable
```

```python
print("Unique O/P Vars: ", data['diagnosis'].unique() ) #Types of O/P variable
print("No of Unique values: ", data['diagnosis'].nunique() )
```

```python
#Dummies in Pandas will Encode Categorical Variables - Binary Encoding as 0 or 1.
data = pd.get_dummies( data=data, drop_first=True)
data.head()
```

```python
print((data.diagnosis_M == 0).sum()) #tumor is Benign
```

```python
print((data.diagnosis_M == 1).sum()) #tumor is Malignant
```

```python
#Split the Dataset into Train and Test after getting X and Y vars:
#INPUT FEATURES:
x = data.iloc[ :, 1:-1].values

#OUTPUT VAR:
y = data.iloc[ :, -1].values
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
print(f"Training set X = {x_train.shape}, Y = {y_train.shape} ")
```

```python
print(f"Test set X = {x_test.shape}, Y = {y_test.shape} ")
```

```python
#Feature Scaling:
from sklearn.preprocessing import StandardScaler


std_sc = StandardScaler()
x_train = std_sc.fit_transform(x_train)
x_test = std_sc.transform(x_test)
```

Simple Logistic Regression for Binary Classification

```python
#import
from sklearn.linear_model import LogisticRegression

#classifier
lr_class = LogisticRegression()
lr_class.fit(x_train, y_train)
```

```python
y_pred = lr_class.predict(x_test)
```

```python
#Check the predictions against the actual y test data and measure the model accuracy:
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, precision_score, recall_score

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
results = pd.DataFrame([['Logistic Regression', accuracy, f1, precision, recall]],
               columns = ['Model', 'Accuracy', 'F1 Score', 'Precision', 'Recall'])
```

```python
results
```

```python
conf_mat = confusion_matrix(y_test, y_pred)
print(conf_mat)
```

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator=lr_class, X=x_train, y=y_train, cv=10)

print("Accuracy is {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation is {:.2f} %".format(accuracies.std()*100))
```
